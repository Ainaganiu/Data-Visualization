```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df1 = 'C:/Users/User/Documents/Python Vizualization/df_Canada.csv'
dfc= pd.read_csv(df1)
```

```python
dfc
```

```python
# Indexing the dataset by Counrty
df = dfc.set_index('Country')
df
```

```python
# Finding Bar chart for Immigrants from Iceland
years = list(map(str,range(1980,2014)))
df_iceland = df.loc['Iceland', years] 
df_iceland.plot(kind='bar')
plt.title('Icelandic Immigrants to Canada fro 1980 to 2013')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')
plt.show()
```

```python
# Plotting Horizontal Bar Chart in Python
df_UK = df.loc['UK',years]
df_UK.plot(kind = 'barh')
plt.title('Icelandic Immigrants to Canada fro 1980 to 2013')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')
plt.show()
```
