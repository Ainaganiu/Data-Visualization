```python
# importing the dataset into pandas dataframe
df11 = 'C:/Users/User/Downloads/Canada+Immigrant.csv'
df1 = pd.read_csv(df11)
df = pd.DataFrame(df1)
```

```python
# printing the read dataset
df
```

```python
# Removing 'Unnamed:0' from the df(dataframe)
df.drop('Unnamed: 0', axis = 1, inplace = True)
```

```python
# Changing the string of column 183
ll = 'UK'
df['OdName'][183] = ll
```

```python
# Finding the total columns in the dataframe
df.columns
```

```python
# # Grouping 1986 to 2013 dataset together
rt = df[['1986', '1987', '1988', '1989', '1990', '1991', '1992', '1993', '1994',
         '1995', '1996', '1997', '1998', '1999', '2000', '2001', '2002', '2003',
         '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012',
         '2013']]
rt
```

```python
# Finding the sum of rt above and adding into df dataframe
rr = rt.sum(axis=1)
df['Total'] = rr
```

```python
# Printing df
df
```

```python
# Renaming odName to Country
df.rename(columns = {'OdName':'Country'}, inplace = True)
```

```python
# Indexing the dataset by Counrty
dfc = df.set_index('Country')
```

```python
# Printing the first 5 dataset in dfc
dfc[0:5]
```

```python
# Sorting by Total in the dataframe
dfc.sort_values(['Total'], ascending= False, axis =0, inplace= True)
dfc.head()
```

```python
# Groping the data in top 5 and transpose them
years = list(map(str,range(1980,2014)))
dfc_top5 = dfc.head()
dfc_top5 = dfc_top5[years].transpose()
```

```python
dfc_top5
```

```python
# Plotting the area plot for top % contries
dfc_top5.plot(kind='area')
plt.title('Immigration trend of top 5 countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
plt.show()
```
