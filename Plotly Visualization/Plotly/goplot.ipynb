```python
df1 = 'C:/Users/User/Downloads/Malrias.xls'
```

```python
df=pd.read_excel(df1)
```

```python
# Now we are finding the total case of malaria that has been reported in the countries from 2008-2016
# We first create a list showing the range of 2006 to 2017
years = list(map(str,range(2006,2017)))

```

```python
# Grouping the incidence of total malarai from 2006-2016
yi=df['Malaria cases reported']
data = [go.Bar(x=years,y=yi)]
layout = go.Layout(title='Malaria Case from 2006 to 2017',
                   xaxis={'title':'Year'},
                   yaxis={'title':'Cases Reported'})
```

```python
fig = go.Figure(data=data,layout=layout)
pyo.plot(fig,filename='BArchart.html')
```
